```python
import pennylane as qml
from pennylane import numpy as np

a = [1, 1, 1, 1, 1]
n = 5
dev = qml.device("lightning.qubit", wires=n + 1)

def oracle(n, a):
    for i in range(n):
        if a[i] == '1':
            qml.CNOT(wires=[i, n])

@qml.qnode(dev)
def bernstein_vasirani_algorythm(n, a):
  #first hadamard
  qml.PauliX(wires=n)
  qml.Hadamard(wires=n)
  for i in range(n):
    qml.Hadamard(wires=i)

  #oracle
  oracle(n, a)

  #second hadamard
  for i in range(n):
    qml.Hadamard(wires=i)

  #measurement
  return  [qml.expval(qml.PauliZ(i)) for i in range(n)]

result = bernstein_vasirani_algorythm(n, a)

print(np.round(result, 6))
```
