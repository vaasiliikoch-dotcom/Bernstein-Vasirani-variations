```python
import cudaq

@cudaq.kernel
def oracle(qc: cudaq.qview, ancilla: cudaq.qubit, a: list[int]):
  for i in range(len(qc)):
    if(a[i] == 1):
      x.ctrl(qc[i], ancilla)


@cudaq.kernel
def main():
  a = [1, 1, 1, 1, 0]
  qc = cudaq.qvector(5)
  ancilla = cudaq.qubit()

  for i in range(5):
    h(qc[i])
  x(ancilla)
  h(ancilla)

  oracle(qc, ancilla, a)

  for i in range(5):
    h(qc[i])
  mz(qc)

counts = cudaq.sample(main)
print(counts)
```
